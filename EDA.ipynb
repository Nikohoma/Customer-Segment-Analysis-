Importing necessary libraries for loading of the dataset.


```python
import pandas as pd
import csv
```

```python
file_path="Mall_Customers_cleaned.csv"
data= pd.read_csv(file_path)
```

Describing the data mathematically.

```python
data.describe()
```

Import necessary libraries for data visualisation.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))   #Setting the area of the chart
sns.histplot(data['Age'],bins=20,kde=True,stat="frequency",color='r') #Plotting the histogram
plt.title('Age Distribution') #Setting the title of the chart
plt.show()
```

```python
plt.figure(figsize=(10,5))   #Setting the area of the chart
sns.ecdfplot(data['Annual Income'],color="g") #Plotting the ecdfplot
plt.title('Annual Income Distribution') #Setting the title of the chart
plt.show()
```

```python
plt.figure(figsize=(10,5))   #Setting the area of the chart
sns.kdeplot(data['Spending Score'],color="g",) #Plotting the kdeplot with the desired color of kde
plt.title('Spending Score Distribution')  #Setting the title of the chart
plt.show()
```

Importing plotly for data visualisation.

```python
import plotly.express as px
```

```python
fig=px.scatter(data,x='Annual Income',y='Spending Score',color='Gender',symbol="Gender",width=900,height=700,title="Annual Income vs Spending Score Gender wise Distribution") #Creating a scatterplot with the markers color depicting Gender
fig.show()
```

```python
#Heatmap showcasing relationship between Annual Income, Spending Score and Age
import plotly.graph_objects as go #importing a plotly module
fig=go.Figure(data=go.Contour(x=data['Annual Income'],y=data['Spending Score'],z=data['Age'])) #Creating a contour style heatmap with the three data input
fig.update_layout(title='Annual Income vs Spending Score Age wise Distribution',width=1000,height=800) #Setting the title and size of the chart
fig.show()
```
